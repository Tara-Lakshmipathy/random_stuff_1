```python
@as_function_node("vacancy_structure")
def CreateVacancyStructure(pr_name: str, element: str) -> CustomVacancyStructure:
    from pyiron_atomistics import Project
    pr = Project(pr_name)
    struct = pr.create.structure.bulk(element, cubic = True)
    vacancy_struct = struct.copy()
    del vacancy_struct[1]

    struct_dataclass = VacancyStructure(structure=vacancy_struct)
    struct_dataclass.CustomSimulationCell.simulation_cell_vectors=[np.round(vacancy_struct.cell[0],4),np.round(vacancy_struct.cell[1],4),np.round(vacancy_struct.cell[2],4)]
    struct_dataclass.CustomSimulationCell.simulation_cell_angles=[np.round(vacancy_struct.cell.cellpar()[3],4),np.round(vacancy_struct.cell.cellpar()[4],4),np.round(vacancy_struct.cell.cellpar()[5],4)]
    struct_dataclass.CustomComposition.chemical_species=vacancy_struct.get_chemical_formula()
    struct_dataclass.CustomCrystalDefect.defect_details_1_a=struct.get_positions()[1]

    return struct_dataclass
```

# Dataclasses

```python
from ase import Atoms
from typing import Optional
import numpy as np
@semantikon_class
@dataclass
class AtomisticSample:
    class_object: Optional[Atoms] = None
    class_uri = AtomicScaleSample
    conceptual_dictionary: u(Optional[dict], label="atomistic_sample", uri=([SimulationCell, ChemicalSpecies, UnitCell, CrystalDefect],
                                                                            [Value, Label, Unit])) = None

    @dataclass
    class SampleSimulationCell:
        simulation_cell_lengths: u(np.ndarray, units="ANGSTROM", label="simulation_cell_lengths", uri=Length) = None
        simulation_cell_vectors: u(np.ndarray, units="ANGSTROM", label="simulation_cell_vectors", uri=SimulationCellVector) = None
        simulation_cell_angles: u(np.ndarray, units="DEGREES", label="simulation_cell_angles", uri=SimulationCellAngle) = None
        simulation_cell_volume: u(float, units="ANGSTROM3", label="simulation_cell_volume", uri=SimulationCellVolume) = None

    @dataclass
    class SampleChemicalSpecies:
        atoms: u(list, label="atoms", uri=Atom) = None
        molecules: u(dict, label="molecules", uri=Molecule) = None

    @dataclass
    class SampleUnitCell:
        lattice_parameter_a: u(float, units="ANGSTROM", label="lattice_parameter_a", uri=LatticeParameter) = None
        lattice_parameter_b: u(float, units="ANGSTROM", label="lattice_parameter_b", uri=LatticeParameter) = None
        lattice_parameter_c: u(float, units="ANGSTROM", label="lattice_parameter_c", uri=LatticeParameter) = None
        lattice_angle_alpha: u(float, units="DEGREES", label="lattice_angle_alpha", uri=LatticeAngle) = None
        lattice_angle_beta: u(float, units="DEGREES", label="lattice_angle_beta", uri=LatticeAngle) = None
        lattice_angle_gamma: u(float, units="DEGREES", label="lattice_angle_gamma", uri=LatticeAngle) = None
        lattice_volume: u(float, units="ANGSTROM3", label="lattice_volume", uri=LatticeVolume) = None
        space_group: u(float, label="space_group", uri=SpaceGroup) = None
        bravais_lattice: u(float, label="bravais_lattice", uri=BravaisLattice) = None
```

```python
import numpy as np
@semantikon_class
@dataclass
class SampleVacancyDefect:
    vacancy_count: u(int, label="defect_count", uri=Vacancy) = None
    vacancy_positions: u(np.ndarray, label="vacancy_positions", uri=Position) = None
    removed_elements: u(list, label="removed_elements", uri=Atom) = None
```

```python
@semantikon_class
@dataclass
class SampleSubstitutionDefect:
    substitution_count: u(int, label="defect_count", uri=Vacancy) = None
    substitution_positions: u(np.ndarray, label="vacancy_positions", uri=Position) = None
    substitution_element: u(np.ndarray, label="vacancy_positions", uri=Position) = None
    substituted_elements: u(list, label="removed_elements", uri=Atom) = None
```

# Parsing stuff

```python
def parse_sim_cell(dc: AtomisticSample.SampleSimulationCell, structure:Atoms):
    dc.simulation_cell_lengths=[np.round(structure.cell.cellpar()[0],4),np.round(structure.cell.cellpar()[1],4),np.round(structure.cell.cellpar()[2],4)]
    dc.simulation_cell_vectors=[np.round(structure.cell[0],4),np.round(structure.cell[1],4),np.round(structure.cell[2],4)]
    dc.simulation_cell_angles=[np.round(structure.cell.cellpar()[3],4),np.round(structure.cell.cellpar()[4],4),np.round(structure.cell.cellpar()[5],4)]
    dc.simulation_cell_volume=np.round(structure.get_volume(),4)
```

```python
def parse_chem_species_atoms(dc: AtomisticSample.SampleChemicalSpecies, structure:Atoms):
    natoms = structure.get_number_of_atoms()
    species_dict = dict(structure.get_number_species_atoms())
    atoms_list = []
    for k in species_dict.keys():
        element = {}
        element["value"] = species_dict[k]
        element["label"] = k
        atoms_list.append(element)
    atoms_list.append({'value': natoms, 'label': 'total_number_atoms'})
    dc.atoms=atoms_list
```

```python
def get_unit_cell_parameters(structure):
    if structure.get_symmetry().spacegroup['InternationalTableSymbol'] == "Im-3m":
        if structure.get_number_of_atoms() == 1:
            structure_parameters = {
                'a': np.round(structure.cell[1][0]*2, 4),
                'alpha': 90.0,
                'beta': 90.0,
                'gamma': 90.0,
                'volume': np.round((structure.cell[1][0]*2)**3, 4),
                'space_group': structure.get_symmetry().spacegroup['InternationalTableSymbol'],
                'bravais_lattice': 'bcc'
            }
        else:
            structure_parameters = {
                'a': np.round(structure.cell[1][1], 4),
                'alpha': 90.0,
                'beta': 90.0,
                'gamma': 90.0,
                'volume': np.round(structure.get_volume(),4),
                'space_group': structure.get_symmetry().spacegroup['InternationalTableSymbol'],
                'bravais_lattice': 'bcc'
            }
    elif structure.get_symmetry().spacegroup['InternationalTableSymbol'] == "Fm-3m":
        if structure.get_number_of_atoms() == 1:
            structure_parameters = {
                'a': np.round(structure.cell[1][0]*2, 4),
                'alpha': 90.0,
                'beta': 90.0,
                'gamma': 90.0,
                'volume': np.round((structure.cell[1][0]*2)**3, 4),
                'space_group': structure.get_symmetry().spacegroup['InternationalTableSymbol'],
                'bravais_lattice': 'fcc'
            }
        else:
            structure_parameters = {
                'a': np.round(structure.cell[1][1], 4),
                'alpha': 90.0,
                'beta': 90.0,
                'gamma': 90.0,
                'volume': np.round(structure.get_volume(),4),
                'space_group': structure.get_symmetry().spacegroup['InternationalTableSymbol'],
                'bravais_lattice': 'fcc'
            }
    elif structure.get_symmetry().spacegroup['InternationalTableSymbol'] == "P6_3/mmc":
        if structure.get_number_of_atoms() == 2:
            structure_parameters = {
                'a': np.round(structure.cell[0][0], 4),
                'c': np.round(structure.cell[2][2], 4),
                'alpha': 90.0,
                'beta': 90.0,
                'gamma': 120.0,
                'volume': np.round(structure.get_volume()*3,4),
                'space_group': structure.get_symmetry().spacegroup['InternationalTableSymbol'],
                'bravais_lattice': 'hcp'
            }
        else:
            structure_parameters = {
                'a': np.round(structure.cell[0][0], 4),
                'c': np.round(structure.cell[2][2], 4),
                'alpha': 90.0,
                'beta': 90.0,
                'gamma': 120.0,
                'volume': np.round(structure.get_volume(),4),
                'space_group': structure.get_symmetry().spacegroup['InternationalTableSymbol'],
                'bravais_lattice': 'hcp'
            }

    return structure_parameters

def parse_unit_cell(dc: AtomisticSample.SampleUnitCell, structure:Atoms):
    structure_parameters = get_unit_cell_parameters(structure)
    dc.lattice_parameter_a=structure_parameters['a']
    if 'b' not in structure_parameters.keys(): 
        dc.lattice_parameter_b=structure_parameters['a']
    else:
        dc.lattice_parameter_b=structure_parameters['b']
    if 'c' not in structure_parameters.keys(): 
        dc.lattice_parameter_c=structure_parameters['a']
    else:
        dc.lattice_parameter_c=structure_parameters['c']
    dc.lattice_angle_alpha=structure_parameters['alpha']
    dc.lattice_angle_beta=structure_parameters['beta']
    dc.lattice_angle_gamma=structure_parameters['gamma']
    dc.lattice_volume=structure_parameters['volume']
    dc.space_group=structure_parameters['space_group']
    dc.bravais_lattice=structure_parameters['bravais_lattice']
```

```python
def populate_cdict_create_structure_node(structure):

    structure_dataclass = AtomisticSample(class_object=structure)

    sim_cell_dataclass = AtomisticSample.SampleSimulationCell()
    parse_sim_cell(sim_cell_dataclass, structure)
    structure_dataclass.conceptual_dictionary = {'simulation_cell': sim_cell_dataclass}

    chem_species_dataclass = AtomisticSample.SampleChemicalSpecies()
    parse_chem_species_atoms(chem_species_dataclass, structure)
    structure_dataclass.conceptual_dictionary['chemical_species'] = chem_species_dataclass

    unit_cell_dataclass = AtomisticSample.SampleUnitCell()
    parse_unit_cell(unit_cell_dataclass, structure)
    structure_dataclass.conceptual_dictionary['unit_cell'] = unit_cell_dataclass

    structure_dataclass.conceptual_dictionary['crystal_defect'] = []

    return structure_dataclass
```

```python
def populate_cdict_repeat_structure_node(repeated_structure_dataclass, structure_dataclass, repeated_structure):
    import copy
    repeated_structure_dataclass.conceptual_dictionary = {}
    
    repeated_sim_cell_dataclass = copy.copy(structure_dataclass.conceptual_dictionary['simulation_cell'])
    parse_sim_cell(repeated_sim_cell_dataclass, repeated_structure)
    repeated_structure_dataclass.conceptual_dictionary['simulation_cell']=repeated_sim_cell_dataclass

    repeated_chem_species_dataclass = copy.copy(structure_dataclass.conceptual_dictionary['chemical_species'])
    parse_chem_species_atoms(repeated_chem_species_dataclass, repeated_structure)
    repeated_structure_dataclass.conceptual_dictionary['chemical_species']=repeated_chem_species_dataclass

    repeated_unit_cell_dataclass = copy.copy(structure_dataclass.conceptual_dictionary['unit_cell'])
    repeated_structure_dataclass.conceptual_dictionary['unit_cell']=repeated_unit_cell_dataclass

    repeated_defect_dataclass_list = copy.copy(structure_dataclass.conceptual_dictionary['crystal_defect'])
    repeated_structure_dataclass.conceptual_dictionary['crystal_defect']=repeated_defect_dataclass_list

    return repeated_structure_dataclass    
```

```python
def populate_cdict_create_vacancy_node(vacancy_structure_dataclass, structure_dataclass, vacancy_structure, vacancy_indices):
    import copy
    vacancy_structure_dataclass.conceptual_dictionary = {}
    
    vacancy_sim_cell_dataclass = copy.copy(structure_dataclass.conceptual_dictionary['simulation_cell'])
    vacancy_structure_dataclass.conceptual_dictionary['simulation_cell']=vacancy_sim_cell_dataclass
    
    vacancy_chem_species_dataclass = copy.copy(structure_dataclass.conceptual_dictionary['chemical_species'])
    parse_chem_species_atoms(vacancy_chem_species_dataclass, vacancy_structure)
    vacancy_structure_dataclass.conceptual_dictionary['chemical_species']=vacancy_chem_species_dataclass
    
    vacancy_unit_cell_dataclass = copy.copy(structure_dataclass.conceptual_dictionary['unit_cell'])
    vacancy_structure_dataclass.conceptual_dictionary['unit_cell']=vacancy_unit_cell_dataclass

    vacancy_defect_dataclass_list = copy.copy(structure_dataclass.conceptual_dictionary['crystal_defect'])

    vacancy_defect_dataclass = SampleVacancyDefect()
    vacancy_defect_dataclass.vacancy_count = len(vacancy_indices)
    positions = []
    for index in vacancy_indices:
        positions.append(structure_dataclass.class_object.get_positions()[index])
    vacancy_defect_dataclass.vacancy_positions = positions
    elements = []
    for index in vacancy_indices:
        elements.append(structure_dataclass.class_object.get_chemical_symbols()[index])
    vacancy_defect_dataclass.removed_elements = elements

    vacancy_defect_dataclass_list.append(vacancy_defect_dataclass)
    vacancy_structure_dataclass.conceptual_dictionary['crystal_defect']=vacancy_defect_dataclass_list

    return vacancy_structure_dataclass
```

```python
def populate_cdict_substitute_impurity_atom_node(subs_structure_dataclass, structure_dataclass, subs_structure, substitute_species, subs_indices):
    import copy
    subs_structure_dataclass.conceptual_dictionary = {}
    
    subs_sim_cell_dataclass = copy.copy(structure_dataclass.conceptual_dictionary['simulation_cell'])
    subs_structure_dataclass.conceptual_dictionary['simulation_cell']=subs_sim_cell_dataclass

    subs_chem_species_dataclass = copy.copy(structure_dataclass.conceptual_dictionary['chemical_species'])
    parse_chem_species_atoms(subs_chem_species_dataclass, subs_structure)
    subs_structure_dataclass.conceptual_dictionary['chemical_species']=subs_chem_species_dataclass

    subs_unit_cell_dataclass = copy.copy(structure_dataclass.conceptual_dictionary['unit_cell'])
    subs_structure_dataclass.conceptual_dictionary['unit_cell']=subs_unit_cell_dataclass

    subs_defect_dataclass_list = copy.copy(structure_dataclass.conceptual_dictionary['crystal_defect'])

    subs_defect_dataclass = SampleSubstitutionDefect()
    subs_defect_dataclass.vacancy_count = len(subs_indices)
    positions = []
    for index in subs_indices:
        positions.append(structure_dataclass.class_object.get_positions()[index])
    subs_defect_dataclass.substitution_positions = positions
    elements = []
    for index in subs_indices:
        elements.append(structure_dataclass.class_object.get_chemical_symbols()[index])
    subs_defect_dataclass.substituted_elements = elements

    subs_defect_dataclass.substitution_element = substitute_species

    subs_defect_dataclass_list.append(subs_defect_dataclass)
    subs_structure_dataclass.conceptual_dictionary['crystal_defect']=subs_defect_dataclass_list

    return subs_structure_dataclass
```

# Nodes and workflow

```python
@as_function_node("structure")
def CreateStructure(pr_name: str, element: str, bravais_lattice: str, a: Optional[int|float], c: Optional[int|float] = None, cubic:bool = False) -> AtomisticSample:
    from pyiron_atomistics import Project
    pr = Project(pr_name)
    if bravais_lattice != 'hcp' and c == None:
        structure = pr.create.structure.bulk(element, crystalstructure=bravais_lattice, cubic=cubic, a=a)
    elif bravais_lattice == 'hcp' and c == None:
        structure = pr.create.structure.bulk(element, crystalstructure=bravais_lattice, cubic=cubic, a=a)
    else:
        structure = pr.create.structure.bulk(element, crystalstructure=bravais_lattice, cubic=cubic, a=a, c=c)

    structure_dataclass = populate_cdict_create_structure_node(structure)

    return structure_dataclass
```

```python
@as_function_node("structure")
def RepeatStructure(structure_dataclass: AtomisticSample, x_repeat: int, y_repeat: int, z_repeat: int) -> AtomisticSample:
    import copy
    repeated_structure_dataclass = copy.copy(structure_dataclass)
    repeated_structure = repeated_structure_dataclass.class_object.repeat([x_repeat, y_repeat, z_repeat])
    repeated_structure_dataclass.class_object = repeated_structure

    repeated_structure_dataclass = populate_cdict_repeat_structure_node(repeated_structure_dataclass, structure_dataclass, repeated_structure)
    
    return repeated_structure_dataclass
```

```python
@as_function_node("structure")
def CreateVacancy(structure_dataclass: AtomisticSample, vacancy_indices: list) -> AtomisticSample:
    import copy
    vacancy_structure_dataclass = copy.copy(structure_dataclass)
    vacancy_structure = vacancy_structure_dataclass.class_object.copy()
    for index in vacancy_indices:
        del vacancy_structure[index]
    vacancy_structure_dataclass.class_object = vacancy_structure
    
    vacancy_structure_dataclass = populate_cdict_create_vacancy_node(vacancy_structure_dataclass, structure_dataclass, vacancy_structure, vacancy_indices)

    return vacancy_structure_dataclass
```

```python
@as_function_node("structure")
def SubstituteImpurityAtom(structure_dataclass: AtomisticSample, substitute_species: str, substitute_magmom: float, subs_indices: list) -> AtomisticSample:
    import copy
    subs_structure_dataclass = copy.copy(structure_dataclass)
    subs_structure = subs_structure_dataclass.class_object.copy()
    for index in subs_indices:
        subs_structure[index] = substitute_species
        subs_structure[index].magmom = substitute_magmom
    subs_structure_dataclass.class_object = subs_structure

    subs_structure_dataclass = populate_cdict_substitute_impurity_atom_node(subs_structure_dataclass, structure_dataclass, subs_structure, substitute_species, subs_indices)

    return subs_structure_dataclass
```
